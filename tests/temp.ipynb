```python
import numpy as np
import pandas as pd

import sys_path # Adds bro to system path
from feature_selection import standardize # Built in helper function

# Objecctive
def f(x):
    """Noise free objective."""
    mean1 = 0.5
    var1 = 2
    mean2 = 5
    var2 = 2
    mean3 = 9
    var3 = 5
    
    return np.exp(-(x-mean1)**2/(2*var1)) + np.exp(-(x-mean2)**2/(2*var2)) + np.exp(-(x-mean3)**2/(2*var3))

# BRO works on discritized domains
X = np.linspace(0,10,500)
noise_free = [f(x) for x in X]
noisy = noise_free + (np.random.random(len(noise_free)) - 0.5)/5

data = pd.DataFrame()
data['x'] = X
data['f'] = noise_free
data['g'] = noisy
```

```python
# Plot the objective and noisy observations

import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['figure.figsize'] = (6,6)
matplotlib.rcParams['font.size'] = 12

plt.plot(data['x'], data['f'], color='black')
plt.scatter(data['x'], data['g'], alpha=0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Optimization Objective')
plt.show()
```

### Expected Improvement

Let's start by defining the BO object for this task EI is wrapped into the EI-TS acquisition function because it is a sequential algorithm.

```python
from bro import BO
from gpytorch.priors import GammaPrior
from pd_utils import to_torch

# Optimize on the unit hypercube with output on [0,100] (like yield optimization)
data = standardize(data)*[1,100,100]

# Initialize a BO object
bo = BO(exindex=data.drop('f', axis=1),
        domain=data.drop(['f','g'], axis=1),
        acquisition_function='EI-TS',
        lengthscale_prior=[GammaPrior(1.1,0.2), 0.2],
        batch_size=1
        )
```

Now we can select random initial samples and run a single iteration of optimization.

```python
# Change initialization batch_size to get some random training data
bo.init_seq.batch_size = 5
bo.init_sample(seed=2, append=True)

# Run an iteration
bo.run()
next_points = bo.obj.get_results(bo.proposed_experiments)
mean = bo.obj.scaler.unstandardize(bo.model.predict(to_torch(bo.obj.domain)))
std = bo.obj.scaler.unstandardize(np.sqrt(bo.model.variance(to_torch(bo.obj.domain))))
ei = bo.obj.scaler.unstandardize(bo.acq.function.ei)
```

And plot the results.

```python
matplotlib.rcParams['figure.figsize'] = (6,6)

plt.figure(1)

# Model mean and standard deviation
plt.subplot(211)
plt.plot(data['x'], data['f'], color='black')
plt.plot(data['x'], mean, label='GP')
plt.fill_between(data['x'], mean-std, mean+std, alpha=0.4)
# Known results and next selected point
plt.scatter(bo.obj.results_input()['x'], bo.obj.results_input()['g'], label='known')
plt.scatter(next_points['x'],next_points['g'], color='red', label='selected')
plt.ylabel('f(x)')
plt.legend(loc='lower right', shadow=True)

# Expected Improvement
plt.subplot(212)
plt.plot(data['x'], ei, color='red')
plt.xlabel('x')
plt.ylabel('EI(x)')
plt.show()
```
